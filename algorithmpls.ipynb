```python
import random
import csv
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import re
```

# Mutation

```python
column_names = [str(i) for i in range(1, 31)]
#test1= pd.read_csv("./single_point_slots_co.csv", names= column_names)
#test1.head()
```

```python
day_swap= pd.read_csv("./day_swap.csv", names= column_names)
day_swap.head()
```

```python
inversion= pd.read_csv("./inversion.csv", names= column_names)
inversion.head()
```

```python
timeslot_swap = pd.read_csv("./timeslot_swap.csv", names= column_names)
timeslot_swap.head()
```

```python
def extract_number(cell):
    match = re.search(r'(\d+\.?\d*)$', cell)
    if match:
        return match.group(1)
    else:
        return None
```

```python
inversion = inversion.applymap(extract_number).astype(float)
timeslot_swap = timeslot_swap.applymap(extract_number).astype(float)
day_swap = day_swap.applymap(extract_number).astype(float)
```

```python
clrs = sns.color_palette("husl", 3)

fig, axes = plt.subplots(1, 3, figsize=(18, 4), sharey=True)

custom_palette = ["#fd7f6f", "#7eb0d5", "#b2e061"]
for i, ax in enumerate(axes.flatten()):
# Select the dataset for the current box plot
    if i == 0:
        dataset = inversion
        title = "Inversion"
    elif i == 1:
        dataset = day_swap
        title = "Day Swap"
    elif i == 2:
        dataset = timeslot_swap
        title = "Timeslot Swap"
    
    # Create the box plot
    sns.boxplot(data=dataset, color=custom_palette[i], ax=ax)
    
    # Adjust y-axis limit to accommodate the maximum value
    ax.set_ylim(0, 5000)
    ax.set_xlabel(title)

# Adjust spacing between subplots
plt.tight_layout()

# Display the plot
plt.show()
```

```python
unpivoted_df1 = pd.melt(inversion, var_name='Generation', value_name='Fitness_inversion')
unpivoted_df2 = pd.melt(day_swap, var_name='Generation', value_name='Fitness_day_swap')
unpivoted_df3 = pd.melt(timeslot_swap, var_name='Generation', value_name='Fitness_timeslot_swap')
unpivoted_df = pd.merge(unpivoted_df1, unpivoted_df2, left_index=True, right_index=True)
unpivoted_df = pd.merge(unpivoted_df, unpivoted_df3, left_index=True, right_index=True)
unpivoted_df.drop(["Generation_x","Generation_y"], axis=1, inplace=True)
unpivoted_df
```

```python
# Reshape the data into a long format
df_long = unpivoted_df.melt(id_vars='Generation', value_vars=['Fitness_inversion', 'Fitness_day_swap', 'Fitness_timeslot_swap'],
                            var_name='Method', value_name='Fitness')
# Create the line plot
sns.lineplot(data=df_long, x='Generation', y='Fitness', hue='Method')

# Set plot title and y-axis limit
plt.title('Fitness by Generation')
plt.ylim(0)

# Display the plot
plt.show()
#plt.savefig('mutation.png')


```
